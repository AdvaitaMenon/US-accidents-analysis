###Start Time

```python
df.Start_Time
```

```python
df.Start_Time= pd.to_datetime(df.Start_Time)
```

```python
df.Start_Time[0]
```

```python
sns.histplot(df.Start_Time.dt.hour,bins=24,kde=False,stat='percent')
```

- A high percentage of accidents occur between  3 pm to 6 pm.

```python
sns.histplot(df.Start_Time.dt.dayofweek,bins=7,kde=False,stat='percent')
```

Is the distribution of accidents by hour same on weekends as on weekdays

```python
sundays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.histplot(sundays_start_time.dt.hour,bins=24,kde=False,stat='percent') 
```

On sundays, the distribution curve is more even.

```python
mondays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 0]
sns.histplot(mondays_start_time.dt.hour,bins=24,kde=False,stat='percent') 
```

```python
sns.displot(df.Start_Time.dt.month,bins=12,kde=False,stat='percent')
```

```python
df_2019=df[df.Start_Time.dt.year==2017]
sns.displot(df_2019.Start_Time.dt.month,bins=12,kde=False,stat='percent')
```
